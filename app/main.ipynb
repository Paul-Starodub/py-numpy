### Task 7.2

````python
# Python code
def unknown_signature(lst1, lst2):
    bitwise_complement = [not x for x in lst1]
    bitwise_and = [x and y for x, y in zip(lst1, lst2)]
    bitwise_or = [x or y for x, y in zip(lst1, lst2)]
    return bitwise_complement, bitwise_and, bitwise_or

list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]
%time unknown_signature(list1, list2)
result_bitwise_complement, result_bitwise_and, result_bitwise_or = unknown_signature(list1, list2)
result_bitwise_complement, result_bitwise_and, result_bitwise_or
````

````python
def unknown_signature(lst1, lst2):
    lst1 = np.array(lst1)
    lst2 = np.array(lst2)

    bitwise_complement = np.logical_not(lst1)
    bitwise_and = np.bitwise_and(lst1, lst2)
    bitwise_or = np.bitwise_or(lst1, lst2)

    return bitwise_complement.tolist(), bitwise_and.tolist(), bitwise_or.tolist()


list1 = [5, 2, 7, 4, 9]
list2 = [3, 6, 1, 8, 10]
%time unknown_signature(list1, list2)
result_bitwise_complement, result_bitwise_and, result_bitwise_or = unknown_signature(list1, list2)
result_bitwise_complement, result_bitwise_and, result_bitwise_or
````

````python
# Python code
def unknown_signature(lst):
    result = []
    for num in lst:
        if num % 2 == 0 and (num % 3 == 0 or num > 10):
            modified_num = (num ^ 7) + 2
            result.append(modified_num)
    return result

my_list = [6, 9, 12, 14, 17, 20]
%time unknown_signature(my_list)
result_python_style = unknown_signature(my_list)
result_python_style
````

````python
def unknown_signature(lst):
    lst = np.array(lst)
    mask = (lst % 2 == 0) & ((lst % 3 == 0) | (lst > 10))
    modified_nums = (lst[mask] ^ 7) + 2
    result = modified_nums.tolist()
    return result

my_list = [6, 9, 12, 14, 17, 20]
%time unknown_signature(my_list)
result_python_style = unknown_signature(my_list)
result_python_style
````
